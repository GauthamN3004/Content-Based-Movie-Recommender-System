# CONTENT BASED RECOMMENDER SYSTEM

```python
import pandas as pd
```

#### READ DATA

```python
df = pd.read_csv("cleaned_movie.csv")
df.head()
```

#### WE WILL ASSUME A USER RATING TO BUILD THE RECOMMENDER ENGINE.

```python
user_rating = {'Toy Story':7,'Get Out(2017 en)':5,'Iron Man(2008 en)':8,'Django Unchained':1,'The Conjuring':9,
              'The Conjuring 2':10,'Titanic(1997 en)':9,'Executive Decision':9,'Leap Year':8,'Zootopia':10,
               'What Happens in Vegas':9,'The Parent Trap(1998 en)':9,'The Shawshank Redemption':3,
              'Baywatch':5,'Johnny English':8,'Inside Out(2015 en)':7,'Toy Story 3':8}
```

#### MAKE A SEPARATE DATAFRAME FOR WATCHED AND UNWATCHED MOVIES

```python
watched = df.loc[df["title"].isin(list(user_rating.keys()))]
not_watched = df.loc[~df["title"].isin(list(user_rating.keys()))]
watched
```

#### A MATRIX FOR GENRE AND LANGUAGE WATCHED BY THE USER

```python
gen_lang_matrix = watched.iloc[:,7:]
gen_lang_matrix.head()
```

#### GENERATING USER-SPECIFIC INFO

AFTER COLLECTING THE MOVIES WATCHED BY THE USER AND THE RATINGS FOR EACH OF THOSE MOVIES.
FIND THE SUM OF EACH COLUMN AND DIVIDE IT BY THE NUMBER OF ROWS.
THIS WILL GIVE THE WEIGHT OF EACH LANGUAGE AND GENRE

```python
user_data = gen_lang_matrix.sum(axis=0)/gen_lang_matrix.shape[0]
user_data
```

```python
cols = gen_lang_matrix.columns
```

#### MULTIPLY THE WEIGHTS TO THE NOT_WATCHED MATRIX

```python
not_watched_gen_lang_matrix = not_watched.iloc[:,7:]
not_watched_gen_lang_matrix.head()
```

```python
not_watched_gen_lang_matrix = not_watched_gen_lang_matrix*user_data
```

```python
not_watched_gen_lang_matrix
```

#### CALCULATE SCORE FOR EACH MOVIE AND FIND THE TOP 10

```python
not_watched_gen_lang_matrix["popularity"] = not_watched["popularity"]/not_watched["popularity"].max()
not_watched_gen_lang_matrix["total_score"] = not_watched_gen_lang_matrix.sum(axis = 1)
not_watched_gen_lang_matrix["id"] = not_watched["id"]

not_watched_gen_lang_matrix.head()
```

```python
recommend = not_watched_gen_lang_matrix[["id","total_score"]]
recommend.head()
```

```python
recommend = recommend.sort_values(by = "total_score",ascending = False)
recommend
```

```python
ids = recommend.iloc[:10,:]["id"]
ids = list(ids)
```

```python
rec = df.iloc[0:0,:]
for i in range(len(ids)):
    rec = rec.append(df.loc[df["id"]==ids[i]],ignore_index = True)
```

```python
rec
```

### HERE WE HAVE THE PREDICTION MADE BASED ON THE USER RATINGS .

### THE PROBLEM WITH THIS APPROACH IS THAT THE RECOMMENDED MOVIES ARE BASED ON WHAT USER HAS ALREADY WATCHED. SO THE USER WILL NOT BE RECOMMENDED  NEW GENRES AND LANGUAGES

### ONE WAY TO SOLVE THIS PROBLEM IS TO USE COLLABORATIVE FILTERING, EITHER USER BASED OR ITEM BASED
